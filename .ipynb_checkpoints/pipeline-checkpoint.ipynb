```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")
train.info()
```

```python
df = pd.concat([train, test], ignore_index=True)
df
```

```python
df['trait'] = df['trait']-1
df.drop(columns=['father', 'mother', 'gender'], inplace=True)
df
```

```python
for i in tqdm(range(1, 15+1)) :
    target = str(i) if i >= 10 else "0"+str(i)
    cols = sorted(df[f"SNP_{target}"].unique().tolist())
    df[f"SNP_{target}"] = df[f"SNP_{target}"].map(lambda x : 0 if x==cols[0] else (1 if x==cols[1] else 2))
df
```

```python
df['class_B'] = df['class'].map(lambda x : 1 if x=='B' else 0)
df['class_C'] = df['class'].map(lambda x : 1 if x=='C' else 0)
df['class'] = df['class'].map(lambda x : 0 if x == 'A' else (1 if x=='B' else 2))

for i in range(1, 15+1) :
    target = str(i) if i >= 10 else "0"+str(i)
    target2 = df.groupby(f'SNP_{target}')['class_B', 'class_C'].sum()
    
    for j in range(3) :
        value = target2['class_B'] / (target2['class_B']+target2['class_C'])
        df[f"SNP_{target}_ratio"] = df[f"SNP_{target}"].map(lambda x : value[0] if x==0 else (value[1] if x==1 else value[2])).fillna(0)    

df
```

```python
target = df.groupby('trait')['class_B', 'class_C'].sum()
target2 = (target['class_B'] / (target['class_B']+target['class_C'])).fillna(0)
df['trait_ratio'] = df[f"trait"].map(lambda x : target2[0] if x==0 else target2[1])
df
```

```python
train2 = df[df['id'].str.contains('TRAIN')].copy().drop(columns=['class_B', 'class_C'])
train2['class'].value_counts()
```

```python
test2 = df[df['id'].str.contains('TEST')].copy().drop(columns=['class_B', 'class_C'])
test2
```

```python
train2.columns.tolist()
```

```python
train2.to_csv("./data/df_train00.csv", index=False)
test2.to_csv("./data/df_test00.csv", index=False)
```

# STEP 02. Modeling 01
- A & not A 구분

```python
import warnings
warnings.filterwarnings(action='ignore')

import os
import gc
import math
import random
import pickle
import pandas as pd
import numpy as np
from tqdm.auto import tqdm

from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split, StratifiedKFold

from xgboost import XGBClassifier, XGBRegressor
from imblearn.over_sampling import SMOTE, SMOTENC
```

```python
xgb_params = {
    'booster': 'gbtree',
    'grow_policy': 'lossguide',
    'max_depth': 0,
    'learning_rate': 0.4,
    # 'n_estimators': 25,
    'reg_lambda': 100,
    'reg_alpha' : 10,
    'subsample': 0.9,
    'num_parallel_tree': 1,
    # 'colsample_bytree' : 0.9,
    'colsample_bynode' : 0.9
    # 'rate_drop': 0.3
}
```

```python
train = pd.read_csv("./data/df_train00.csv")
test = pd.read_csv("./data/df_test00.csv")
train.shape, test.shape
```

```python
def calc_f1_score(y_hat, data):
    y_true = data.get_label()
    y_hat = np.round(y_hat)
    return 'f1', f1_score(y_true, y_hat, average="macro")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
random_seed = 6327

y = (train['class'].values > 0).astype(int)
X = train.drop(['id', 'class'], axis=1)
X.iloc[:,:15] = X.iloc[:,:16].astype('category')
X_test = test.drop(['id', 'class'], axis=1)
X_test.iloc[:,:15] = X_test.iloc[:,:16].astype('category')

skf = StratifiedKFold(n_splits=2, shuffle=True, random_state=random_seed)

oof_val_preds1 = np.zeros((X.shape[0], 2))
oof_val_preds2 = np.zeros((X.shape[0], 2))
oof_val_preds3 = np.zeros((X.shape[0], 2))
oof_test_preds1 = np.zeros((X_test.shape[0], 2))
oof_test_preds2 = np.zeros((X_test.shape[0], 2))
oof_test_preds3 = np.zeros((X_test.shape[0], 2))

# OOF
for fold, (train_idx, valid_idx) in enumerate(skf.split(X, y)):

    # print('#'*30, f'Fold [{fold+1}/{skf.n_splits}]', '#'*30)

    # train, valid data 설정
    X_train, y_train = X.iloc[train_idx], y[train_idx]
    X_valid, y_valid = X.iloc[valid_idx], y[valid_idx]
    
    smote = SMOTENC(categorical_features=[x for x in range(16)], random_state=random_seed)
    X_train, y_train = smote.fit_resample(X_train, y_train)

    # 불균형 데이터 가중치 조정 값 => 음성(0) 타깃값 개수 / 양성(1) 타깃값 개수
    _, counts = np.unique(np.array(y_train), return_counts=True)
    scale_weight = counts[0] / counts[1]

    # XGBoost 모델 훈련
    xgb_model = XGBClassifier(
        **xgb_params,
        tree_method='gpu_hist',
        predictor='gpu_predictor',
        random_state=random_seed,
        n_jobs=-1,
        enable_categorical=True
    )
    
    rf = RandomForestClassifier()
    lr = LogisticRegression(C=100)
    
    xgb_model.fit(X_train, y_train, eval_set=[(X_valid, y_valid)], eval_metric=calc_f1_score, verbose=False)
    rf.fit(X_train, y_train)
    lr.fit(X_train, y_train)

    oof_test_preds1 += xgb_model.predict_proba(X_test) / skf.n_splits
    oof_test_preds2 += rf.predict_proba(X_test) / skf.n_splits
    oof_test_preds3 += lr.predict_proba(X_test) / skf.n_splits
    oof_val_preds1[valid_idx] += xgb_model.predict_proba(X_valid)
    oof_val_preds2[valid_idx] += rf.predict_proba(X_valid)
    oof_val_preds3[valid_idx] += lr.predict_proba(X_valid)
    
    if fold == 1 :
        pred = xgb_model.predict(X_test)
        break
    
    #model save
    # xgb_model.save_model(f'./models/new_xgb_{skf.n_splits}_{fold}.json')
    del [[X_train, y_train, X_valid, y_valid, xgb_model]]
    gc.collect()

#     model score check
preds = np.argmax(oof_val_preds1, axis=1)
preds2 = np.argmax(oof_val_preds2, axis=1)
preds3 = np.argmax(oof_val_preds3, axis=1)
print(f1_score(y, preds, average="macro"))
print(f1_score(y, preds2, average="macro"))
print(f1_score(y, preds3, average="macro"))

# # save OOF test preds
# np.save(f'./results/new_{skf.n_splits}_oof_test_preds.npy', oof_test_preds[:, 1])
```

## Not A에 대한 확률예측값을 새로운 변수로 추가

```python
train['preds_xgb'] = pd.Series(oof_val_preds1[:, 0])
train['preds_rf'] = pd.Series(oof_val_preds2[:, 0])
train['preds_lr'] = pd.Series(oof_val_preds3[:, 0])
test['preds_xgb'] = pd.Series(oof_test_preds1[:,0])
test['preds_rf'] = pd.Series(oof_test_preds2[:,0])
test['preds_lr'] = pd.Series(oof_test_preds3[:,0])

train[(train['preds_xgb'] >= 0.5) | (train['preds_rf'] >= 0.5) | (train['preds_lr'] >= 0.5)][train['class']==1]
```

# STEP 03. MODELING 02
- B와 C를 구분

```python
import warnings
warnings.filterwarnings(action='ignore')

import os
import gc
import math
import random
import pickle
import pandas as pd
import numpy as np
from tqdm.auto import tqdm

from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler

from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE, SMOTENC
import torch
```

```python
ae = pd.read_csv("./data/ae_values.csv")
target_idx = train[train['class'] != 0].index.tolist()
train2 = train.iloc[target_idx].copy().reset_index(drop=True)
train2['class'] = train2['class'] -1

train2 = pd.concat([train2, ae[:len(train2)]], axis=1)
test2 = pd.concat([test, ae[len(train2):].reset_index(drop=True)], axis=1)
train2.info()
```

```python
train2.describe()
```

```python
def calc_f1_score(y_hat, data):
    y_true = data.get_label()
    y_hat = np.round(y_hat)
    return 'f1', f1_score(y_true, y_hat, average="macro")
```

```python
xgb_params = {
    'booster': 'gbtree',
    'grow_policy': 'lossguide',
    'max_depth': 0,
    'learning_rate': 0.4,
    # 'n_estimators': 25,
    'reg_lambda': 100,
    'reg_alpha' : 10,
    'subsample': 0.9,
    'num_parallel_tree': 1,
    # 'colsample_bytree' : 0.9,
    'colsample_bynode' : 0.9
    # 'rate_drop': 0.3
}
```

```python
train2.info()
```

```python
train2['class'].value_counts()
```

```python
random_seed = 33070

y = torch.LongTensor(train2['class'].values)
X = train2.drop(['id', 'class', 'trait', 'preds_xgb', 'preds_rf', 'preds_lr'], axis=1).to_numpy()
X_test = test2.drop(['id', 'class', 'trait', 'preds_xgb', 'preds_rf', 'preds_lr'], axis=1).to_numpy()

skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=random_seed)

oof_val_preds = np.zeros((X.shape[0], 2))
oof_val_preds2 = np.zeros((X.shape[0], 2))
oof_val_preds3 = np.zeros((X.shape[0], 2))
oof_test_preds = np.zeros((X_test.shape[0], 2))

# OOF
ests = [12, 23, 17]
volumes = [1000, 1000, 1200]
for fold, (train_idx, valid_idx) in enumerate(skf.split(X, y)):

    print('#'*30, f'Fold [{fold+1}/{skf.n_splits}]', '#'*30)

    # train, valid data 설정
    X_train, y_train = X[train_idx], y[train_idx]
    X_valid, y_valid = X[valid_idx], y[valid_idx]
    
    strategy = {0:volumes[fold], 1:volumes[fold]}
    # strategy = 'auto'
    smote = SMOTE(random_state=random_seed, k_neighbors=4, sampling_strategy=strategy)
    X_train, y_train = smote.fit_resample(X_train, y_train)

    # 불균형 데이터 가중치 조정 값 => 음성(0) 타깃값 개수 / 양성(1) 타깃값 개수
    _, counts = np.unique(np.array(y_train), return_counts=True)
    scale_weight = counts[0] / counts[1]

    # XGBoost 모델 훈련
    xgb_model = XGBClassifier(
        **xgb_params,
        tree_method='gpu_hist',
        predictor='gpu_predictor',
        random_state=random_seed,
        n_estimators=ests[fold],
        n_jobs=-1
    )
    
    rf = RandomForestClassifier()
    lr = LogisticRegression(C=10)
    
    xgb_model.fit(X_train, y_train, eval_set=[(X_valid, y_valid)], eval_metric=calc_f1_score, verbose=True)
    rf.fit(X_train, y_train)
    lr.fit(X_train, y_train)

    oof_test_preds += xgb_model.predict_proba(X_test) / skf.n_splits
    oof_val_preds[valid_idx] += xgb_model.predict_proba(X_valid)
    oof_val_preds2[valid_idx] += rf.predict_proba(X_valid)
    oof_val_preds3[valid_idx] += lr.predict_proba(X_valid)
    
    #model save
    # xgb_model.save_model(f'./models/new_xgb_{skf.n_splits}_{fold}.json')
    del [[X_train, y_train, X_valid, y_valid, xgb_model]]
    gc.collect()

#     model score check
preds = np.argmax(oof_val_preds, axis=1)
preds2 = np.argmax(oof_val_preds2, axis=1)
preds3 = np.argmax(oof_val_preds3, axis=1)
print(f1_score(y, preds, average="macro"))
print(f1_score(y, preds2, average="macro"))
print(f1_score(y, preds3, average="macro"))

# # save OOF test preds
# np.save(f'./results/new_{skf.n_splits}_oof_test_preds.npy', oof_test_preds[:, 1])
```
