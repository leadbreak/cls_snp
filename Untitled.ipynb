```python
import warnings
warnings.filterwarnings(action='ignore')

import os
import gc
import math
import random
import pickle
import pandas as pd
import numpy as np
from tqdm.auto import tqdm

from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split, StratifiedKFold

from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
```

```python
from xgboost import XGBClassifier, XGBRegressor
from imblearn.over_sampling import SMOTE

xgb_params = {
    'booster': 'gbtree',
    'grow_policy': 'lossguide',
    'max_depth': 0,
    'learning_rate': 0.4,
    'n_estimators': 30,
    'reg_lambda': 100,
    'subsample': 0.9,
    'num_parallel_tree': 1,
    # 'rate_drop': 0.3
}
```

```python
train = pd.read_csv("./data/df_train5.csv")
test = pd.read_csv("./data/df_test5.csv")
train.shape, test.shape
```

```python
train = train[train['class']!=0].copy().reset_index(drop=True).drop(columns=['trait'])
test.drop(columns=['trait'], inplace=True)
train['class'] = train['class'] - 1
train['class'].value_counts()
```

```python
train.to_csv("./data/df_train6.csv", index=False)
test.to_csv("./data/df_test6.csv", index=False)
```

```python
def calc_f1_score(y_hat, data):
    y_true = data.get_label()
    y_hat = np.round(y_hat)
    return 'f1', f1_score(y_true, y_hat, average="macro")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
random_seed = 0

y = train['class'].values
X = train.drop(['id', 'class'], axis=1).to_numpy()
X_test = test.drop(['id', 'class'], axis=1).to_numpy()

skf = StratifiedKFold(n_splits=2, shuffle=True, random_state=random_seed)

oof_val_preds = np.zeros((X.shape[0], 2))
oof_val_preds2 = np.zeros((X.shape[0], 2))
oof_val_preds3 = np.zeros((X.shape[0], 2))
oof_test_preds = np.zeros((X_test.shape[0], 2))

# OOF
for fold, (train_idx, valid_idx) in enumerate(skf.split(X, y)):

    print('#'*30, f'Fold [{fold+1}/{skf.n_splits}]', '#'*30)

    # train, valid data 설정
    X_train, y_train = X[train_idx], y[train_idx]
    X_valid, y_valid = X[valid_idx], y[valid_idx]
    
    # strategy = {0:114, 1:114}
    strategy = 'auto'
    smote = SMOTE(random_state=random_seed, k_neighbors=4, sampling_strategy=strategy)
    X_train, y_train = smote.fit_resample(X_train, y_train)

    # 불균형 데이터 가중치 조정 값 => 음성(0) 타깃값 개수 / 양성(1) 타깃값 개수
    _, counts = np.unique(np.array(y_train), return_counts=True)
    scale_weight = counts[0] / counts[1]

    # XGBoost 모델 훈련
    xgb_model = XGBClassifier(
        **xgb_params,
        tree_method='gpu_hist',
        predictor='gpu_predictor',
        random_state=random_seed,
        n_jobs=-1
    )
    
    rf = RandomForestClassifier()
    lr = LogisticRegression(C=100)
    
    xgb_model.fit(X_train, y_train, eval_set=[(X_valid, y_valid)], eval_metric=calc_f1_score, verbose=True)
    rf.fit(X_train, y_train)
    lr.fit(X_train, y_train)

    oof_test_preds += xgb_model.predict_proba(X_test) / skf.n_splits
    oof_val_preds[valid_idx] += xgb_model.predict_proba(X_valid)
    oof_val_preds2[valid_idx] += rf.predict_proba(X_valid)
    oof_val_preds3[valid_idx] += lr.predict_proba(X_valid)
    
    #model save
    # xgb_model.save_model(f'./models/new_xgb_{skf.n_splits}_{fold}.json')
    del [[X_train, y_train, X_valid, y_valid, xgb_model]]
    gc.collect()

#     model score check
preds = np.argmax(oof_val_preds, axis=1)
preds2 = np.argmax(oof_val_preds2, axis=1)
preds3 = np.argmax(oof_val_preds3, axis=1)
print(f1_score(y, preds, average="macro"))
print(f1_score(y, preds2, average="macro"))
print(f1_score(y, preds3, average="macro"))

# # save OOF test preds
# np.save(f'./results/new_{skf.n_splits}_oof_test_preds.npy', oof_test_preds[:, 1])
```
